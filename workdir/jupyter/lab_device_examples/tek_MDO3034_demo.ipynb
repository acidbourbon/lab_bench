```python
import numpy as np
from matplotlib import pyplot as plt

import MDO3034 as scope
scope.init("192.168.0.34")
```

```python
##################################################
##           configure Tek Scope                ##
##################################################


s = {
    "test"   :"CH2", 
    "ref"    :"CH1", 
}

```

```python
##################################################
##              capture waveforms               ##
##################################################

time, wfm = scope.capture_waveforms([
    s["ref"],
    s["test"]
  ],
    average=1
)

##################################################
##                     plot                     ##
##################################################


plt.plot(time, wfm[s["test"]], "c" ,label="test")
plt.plot(time, wfm[s["ref"]]/10., "y" ,label="ref")
plt.legend()
plt.xlabel("time (s)")
plt.ylabel("voltage (V)")
plt.show()
```
