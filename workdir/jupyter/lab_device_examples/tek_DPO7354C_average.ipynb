```python
import numpy as np
from matplotlib import pyplot as plt

import DPO7354C as scope
scope.init("192.168.0.219")
from sigproc_kit import *
```

```python
##################################################
##           configure Tek Scope                ##
##################################################


s = {
    "test"   :"CH3", 
}

```

```python
##################################################
##              capture waveforms               ##
##################################################

time, wfm = scope.capture_waveforms([
    s["test"],
  ],
    average=100
)

##################################################
##                     plot                     ##
##################################################

nobl = remove_baseline(wfm[s["test"]])

plt.plot(time, nobl, "c" ,label="amp unpowered")
plt.legend()
plt.xlabel("time (s)")
plt.ylabel("voltage (V)")
plt.show()
```

```python
##################################################
##              capture waveforms               ##
##################################################

time, wfm = scope.capture_waveforms([
    s["test"],
  ],
    average=100
)

##################################################
##                     plot                     ##
##################################################

nobl2 = remove_baseline(wfm[s["test"]])

plt.plot(time, nobl, "c" ,label="amp unpowered")
plt.plot(time, nobl2, "c" ,label="amp powered")
plt.legend()
plt.xlabel("time (s)")
plt.ylabel("voltage (V)")
plt.show()
```

```python

plt.plot(time*1e9, nobl *1e3  ,label="amp unpowered")
plt.plot(time*1e9, nobl2*1e3  ,label="amp powered")
plt.legend()
plt.xlabel("time (ns)")
plt.ylabel("voltage (mV)")
plt.title("impedance matching to 330R")
plt.show()
```
