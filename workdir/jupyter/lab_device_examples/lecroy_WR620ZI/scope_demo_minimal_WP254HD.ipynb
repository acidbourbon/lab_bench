```python
import numpy as np
from matplotlib import pyplot as plt

import WP254HD as lecroy
lecroy.init("192.168.0.103")
```

```python
##################################################
##           configure LeCroy Scope             ##
##################################################

#lecroy.clear_all()

s = {
    "test"   :"C3", 
}

```

```python
##################################################
##              capture waveforms               ##
##################################################

time, wfm = lecroy.capture_waveforms([
    s["test"],
  ],
    average=1
)

plt.plot(time, wfm[s["test"]], "c" ,label="test")
plt.legend()
plt.xlabel("time (s)")
plt.ylabel("voltage (V)")
plt.show()
```
