```python
import numpy as np
from matplotlib import pyplot as plt

import WR620ZI as lecroy
lecroy.init("192.168.0.208")
```

```python
##################################################
##           configure LeCroy Scope             ##
##################################################

#lecroy.clear_all()

s = {
    "X"   :"C2", 
    "Y"   :"C3", 
}

```

```python
##################################################
##              capture waveforms               ##
##################################################

time, wfm = lecroy.capture_waveforms([
    s["X"],
    s["Y"],
  ],
    average=1
)

```

```python
##################################################
##                     plot                     ##
##################################################


plt.plot(time, wfm[s["X"]], "r" ,label="test", alpha=0.33)
plt.plot(time, wfm[s["Y"]], "b" ,label="test", alpha=0.33)
plt.legend()
plt.xlabel("time (s)")
plt.ylabel("voltage (V)")
plt.show()
```
