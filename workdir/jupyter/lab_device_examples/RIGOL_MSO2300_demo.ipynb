```python
##################################################
##           configure Tek Scope                ##
##################################################


s = {
    "test"   :"2", 
    "ref"    :"1", 
}

```

```python
##################################################
##              capture waveforms               ##
##################################################

time, wfm = scope.capture_waveforms([
    s["test"],
    s["ref"],
  ],
    average=1
)

##################################################
##                     plot                     ##
##################################################


plt.plot(time, wfm[s["test"]], "c" ,label="test")
plt.plot(time, wfm[s["ref"]], "y" ,label="ref")
plt.legend()
plt.xlabel("time (s)")
plt.ylabel("voltage (V)")
plt.grid("True")
plt.show()
```
