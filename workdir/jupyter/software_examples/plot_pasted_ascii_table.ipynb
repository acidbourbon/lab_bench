```python
a = np.loadtxt(StringIO(a.replace("|"," ")))
b = np.loadtxt(StringIO(b.replace("|"," ")))
```

```python
thresh = a[:,0]
ch1a   = a[:,1]
ch2a   = a[:,2]
ch1b   = b[:,1]
ch2b   = b[:,2]
```

```python
plt.plot(thresh,ch1a,label="CH1 - no power filter")
plt.plot(thresh,ch2a,label="CH2 - no power filter")
plt.plot(thresh,ch1b,label="CH1 - with power filter")
plt.plot(thresh,ch2b,label="CH2 - with power filter")
plt.xlabel("threshold")
plt.ylabel("scaler rate")
plt.legend()
plt.title("S-Curve PDC discrete")
plt.yscale("log")
plt.show()
```
