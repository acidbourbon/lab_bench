```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Create data for the trajectory
t = np.linspace(0, 10, 100)  # Time variable
x = np.sin(t)                 # X coordinates
y = np.cos(t)                 # Y coordinates
z = t                         # Z coordinates (time)


# Step 4: Show the plot
```

```python

%matplotlib notebook
# Step 2: Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Step 3: Plot the trajectory
ax.plot(x, y, z, label='3D Trajectory', color='b')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Trajectory Plot')
ax.legend()

plt.show()
```
