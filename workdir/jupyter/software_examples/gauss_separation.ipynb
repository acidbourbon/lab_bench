```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
##################################################
##          create arbitrary waveform           ##
##################################################


def gauss(x, **kwargs):
  mu = kwargs.get("mu",0)
  sigma = kwargs.get("sigma",1)
  ## default amplitude generates bell curve with area = 1
  A = kwargs.get("A",1./(sigma*(2.*np.pi)**0.5))
  return A*np.exp(-(x-mu)**2/(2.*sigma**2))



period = 100e-9

x=np.arange(0,period,0.1e-9)



for i in range(0,6):
  sigma=5e-9
  mu1 = 30e-9
  mu2 = 30e-9 + i*sigma
  
  y=    gauss(x,sigma=sigma,mu=mu1,A=100e-3)+    gauss(x,sigma=sigma,mu=mu2,A=100e-3)
  
  plt.plot(x*1e9,y,label="mu2-mu1 = {:d} sigma".format(i))

plt.legend()
plt.xlabel("time (ns)")
plt.ylabel("voltage (V)")
plt.show()

```
