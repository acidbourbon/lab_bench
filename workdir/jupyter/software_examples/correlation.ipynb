```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def pearson(x,y):
  my_sum = 0
  mean_x = np.mean(x)
  mean_y = np.mean(y)
  for i in range(0,len(x)):
    my_sum += (x[i]-mean_x)*(y[i]-mean_y)
  return (my_sum/(len(x)))/(np.std(x)*np.std(y))
```

```python
size = 1000

x = np.random.normal(size=size)
y = -x + 0.5*np.random.normal(size=size)
#y = np.random.normal(size=size)
```

```python
plt.scatter(x,y,alpha=1,s=1)
#mask = (y>1) & (y<1.2)
#plt.scatter(x[mask],y[mask],alpha=1,s=1)
plt.show()

print("pearson coefficient of x and y: {:3.3f}".format(pearson(x,y)))
```

```python
from scipy.optimize import curve_fit

def func(x, a, b):
    return a*x + b

p0 = [2,-0.5]
popt, pcov = curve_fit(func, x, y, p0 = p0)

plt.scatter(x,y,alpha=0.5,s=1)
#plt.plot(x,func(x,*p0),"r")
plt.plot(x,func(x,*popt),"g")
plt.show()

print(popt)
```
