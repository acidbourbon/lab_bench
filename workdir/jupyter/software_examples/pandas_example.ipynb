```python
x = np.linspace(0,10,100)
y = np.sin(x)

df = pd.DataFrame({
    "x" : x,
    "y" : y
})
```

```python
plt.plot(x,y)
plt.show()

plt.plot(df)
plt.show()

plt.plot(df["x"],df["y"])
plt.show()
```
