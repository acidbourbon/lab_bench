```python
from sympy import *
from sympy.plotting import plot3d, plot_parametric
```

```python
%matplotlib notebook
```

```python
x, y = symbols('x y')

plot3d(log(x+1j*y), (x, -1, 1), (y, -1, 1), title="Re(complex logarithm)",xlabel="X",ylabel="Y")
```

```python

plot3d(im(log(x+1j*y)), (x, -1, 1), (y, -1, 1), title="Im(complex logarithm)",xlabel="X",ylabel="Y")
```
