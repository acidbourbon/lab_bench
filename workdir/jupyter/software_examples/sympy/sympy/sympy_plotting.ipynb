```python
from sympy import symbols, cos, sin

from sympy.plotting import plot3d_parametric_surface
```

```python

phi, theta = symbols('phi, theta')
plot3d_parametric_surface(cos(phi)*sin(theta), sin(phi)*sin(theta), cos(theta),

    (phi, 0, 3*pi/2), (theta, 0, pi))
```

```python
from sympy.plotting import plot3d_parametric_line

u = symbols('u')
plot3d_parametric_line(cos(u), sin(u), u, (u, -5, 5))
```

```python
from sympy.plotting import plot3d_parametric_line

phi, theta = symbols('phi, theta')

phi = 20*theta

plot3d_parametric_line(  cos(phi)*sin(theta), sin(phi)*sin(theta), cos(theta),
    (theta, 0, pi))
```

```python
from sympy import plot_implicit, symbols, Eq, And

x, y = symbols('x y')

p1 = plot_implicit(Eq(x**2 + y**2, 16), (x, -5, 5), (y, -5, 5),show=False)
p2 = plot_implicit(Eq(x**2 + y**2, 10), (x, -5, 5), (y, -5, 5),show=False)
p1.extend(p2)
p1.show()
```

```python
from sympy import plot_implicit, symbols, Eq, And

x, y = symbols('x y')

p1 = plot_implicit(Eq(x**2 + y**2, 16), (x, -5, 5), (y, -5, 5),show=False)
for i in range(1,20):
    p2 = plot_implicit(Eq(x**2 + y**2, 16-i), (x, -5, 5), (y, -5, 5),show=False)
    p1.extend(p2)
p1.show()
```

```python
# implicit plot areas

plot_implicit(y > x**2)
```

```python
# implicit plot areas


plot_implicit(And(y > x, y> 0))
```
