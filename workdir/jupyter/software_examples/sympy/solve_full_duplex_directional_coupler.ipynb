```python


R1, R2, R3, R4, R5, R6, T1, F, Z0 = symbols("R_1 R_2 R_3 R_4 R_5 R_6 T_1 F Z_0")

##################################################
##               general solution               ##
##################################################


# F is the gain going from L to R
# also the voltage divider ratio of the left leg


equations = [
    T1 - Z0,
    1/( 1/R6  + 1/(R5+  1/(1/R2 + 1/(R3+ 1/(1/R4+1/R1))))) - Z0, # everything needs to look Z0 from the right
    (1/(1/R6 + 1/T1))/(R5+1/(1/R6 + 1/T1)) - F, # gain from L to R
    (1/(1/R1+1/R4)) / (R3+ 1/(1/R1+1/R4))  - F, # gain from L to L'
    R1 - 220,
    R2 - 50,
    R3 - 82,
    R4 - 220,
    Z0 - 50
    ]
unknowns  = [R1, R2, R3, R4, R5, R6, T1, F, Z0]

solutions = solve(equations,unknowns)
print(unknowns)
# Convert solutions to floating point
pprint(solutions)

print ("  ---- numeric ----  ")

for i in range(len(solutions)):
    for j in range(len(solutions[i])):
        print ("{}: {}".format( unknowns[j], N(solutions[i][j],5)))
    
    print()
```

```python


R1, R2, R3, R4, R5, R6, T1, F, Z0 = symbols("R_1 R_2 R_3 R_4 R_5 R_6 T_1 F Z_0")

##################################################
##               general solution               ##
##################################################


# F is the gain going from L to R
# also the voltage divider ratio of the left leg

# fraction is already determined by the left leg
# R5 is determined by left leg (fraction) and R6

equations = [
    T1 - Z0,
    1/( 1/R6  + 1/(R5+  1/(1/R2 + 1/(R3+ 1/(1/R4+1/R1))))) - Z0, # everything needs to look Z0 from the right
    (1/(1/R6 + 1/T1))/(R5+1/(1/R6 + 1/T1)) - F, # gain from L to R
    (1/(1/R1+1/R4)) / (R3+ 1/(1/R1+1/R4))  - F, # gain from L to L'
    R1 - 270,
    R2 - 50,
    R3 - 100,
    R4 - R1,
    Z0 - 50,
    ]
unknowns  = [R1, R2, R3, R4, R5, R6, T1, F, Z0]

solutions = solve(equations,unknowns)
print(unknowns)
# Convert solutions to floating point
pprint(solutions)

print ("  ---- numeric ----  ")

for i in range(len(solutions)):
    for j in range(len(solutions[i])):
        print ("{}: {}".format( unknowns[j], N(solutions[i][j],5)))
    
    print()
```
