```python
from sympy.plotting import plot3d_parametric_line
from sympy import *

%matplotlib notebook

phi, theta = symbols('phi, theta')

# this is gold!!!
phi = 10*theta

plot3d_parametric_line(  cos(phi)*sin(theta), sin(phi)*sin(theta), cos(theta),
    (theta, 0, pi))
    
```

```python
from sympy import symbols, cos, sin

from sympy.plotting import plot3d_parametric_surface

phi, theta = symbols('phi, theta')
plot3d_parametric_surface(cos(phi)*sin(theta), sin(phi)*sin(theta), cos(theta),

    (phi, 0, 3.5*pi/2), (theta, 0, pi))
```
