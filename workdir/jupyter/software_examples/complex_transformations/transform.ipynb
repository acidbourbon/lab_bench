```python
import numpy as np

import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
  
# Read Images 
img = mpimg.imread('test_very_big.png') 
  
# Output Images 
plt.imshow(img) 
plt.show()

img = (1-img)*1

height,width, dummy = img.shape
```

```python
def plot_c(z):
  plt.scatter(np.real(z),np.imag(z),s=0.1)
  #plt.show()
```

```python
x = []
y = []

scale = 1/(width/2)

for i in range(height):
  for j in range(width):
    if img[i,j,0] > 0:
      x += [j]
      y += [-i]
x = scale*(np.array(x) - width/2)
y = scale*(np.array(y) + height/2)
z = x+1j*y

plot_c(z)
```

```python
plot_c(z*(2+1j))
```

```python
plot_c(z*np.exp(-45/360 * 2*np.pi*1j))
```

```python
plot_c(np.sin(z))
```

```python
plot_c(np.cos(z))
```

```python
plot_c(np.log(z))
```

```python
plt.scatter(np.exp(np.real(np.log(z))),np.imag(np.log(z)),s=0.1)
```
