```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def dB(x):
    return 20*np.log10(abs(x))
```

```python
##################################################
##                RC integrator                 ##
##################################################


omega_min = 0.1
omega_max = 100
omega_steps = 100
x = np.linspace(np.log10(omega_min),np.log10(omega_max),omega_steps)
omega = 10**x

s = omega*1j

RC = 1

plt.plot(omega,dB(1/(s)),label="pure_integrator")
plt.plot(omega,dB(1/(1+RC*s)),label="RC")
plt.xlabel("omega")
plt.ylabel("dB")
plt.xscale("log")
plt.yscale("linear")
plt.legend()
plt.show()
```

```python
##################################################
##                RC integrator                 ##
##################################################


omega_min = 0.1
omega_max = 100
omega_steps = 100
x = np.linspace(np.log10(omega_min),np.log10(omega_max),omega_steps)
omega = 10**x

s = omega*1j

RC = 0.1

plt.plot(omega,dB(10*1/(s)),label="10*pure_integrator")
plt.plot(omega,dB(1/(1+RC*s)),label="RC")
plt.xlabel("omega")
plt.ylabel("dB")
plt.xscale("log")
plt.yscale("linear")
plt.legend()
plt.show()
```
