```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def dB(x):
    return 20*np.log10(abs(x))
```

```python
##################################################
##                RC integrator                 ##
##################################################


omega_min = 0.1
omega_max = 100
omega_steps = 100
x = np.linspace(np.log10(omega_min),np.log10(omega_max),omega_steps)
omega = 10**x

s = omega*1j

RC = 1

plt.plot(omega,dB(1/(s)),label="pure_integrator")
plt.plot(omega,dB(1/(1+RC*s)),label="RC")
plt.xlabel("omega")
plt.ylabel("dB")
plt.xscale("log")
plt.yscale("linear")
plt.legend()
plt.show()
```

```python
##################################################
##                RC integrator                 ##
##################################################


omega_min = 0.1
omega_max = 100
omega_steps = 100
x = np.linspace(np.log10(omega_min),np.log10(omega_max),omega_steps)
omega = 10**x

s = omega*1j

RC = 0.1

plt.plot(omega,dB(10*1/(s)),label="10*pure_integrator")
plt.plot(omega,dB(1/(1+RC*s)),label="RC")
plt.xlabel("omega")
plt.ylabel("dB")
plt.xscale("log")
plt.yscale("linear")
plt.legend()
plt.show()
```

```python
##################################################
##                  two pole RC                 ##
##################################################


omega_min = 0.1
omega_max = 1000
omega_steps = 100
x = np.linspace(np.log10(omega_min),np.log10(omega_max),omega_steps)
omega = 10**x

s = omega*1j

# first pole
RCa = 1

# second pole
RCb = 0.01

plt.plot(omega,dB(1/(s)),"--",label="pure_integrator")
plt.plot(omega,dB(1/(1+RCa*s)),label="RCa")
plt.plot(omega,dB(1/(1+RCa*s)*1/(1+RCb*s)),label="RCa->RCb")
plt.xlabel("omega")
plt.ylabel("dB")
plt.xscale("log")
plt.yscale("linear")
plt.legend()
plt.show()
```

```python
##################################################
##                   RC vs CR                   ##
##################################################


omega_min = 0.1
omega_max = 10
omega_steps = 100
x = np.linspace(np.log10(omega_min),np.log10(omega_max),omega_steps)
omega = 10**x

s = omega*1j

RC = 1

plt.plot(omega,dB(1/(s)),label="integrator")
plt.plot(omega,dB(1/(1+RC*s)),label="RC")
plt.plot(omega,dB(s),label="differentiator")
plt.plot(omega,dB((RC*s)/(1+RC*s)),label="CR")
plt.xlabel("omega")
plt.ylabel("dB")
plt.xscale("log")
plt.yscale("linear")
plt.legend()
plt.show()
```

$$ H_{RC}(s) = V_{out}/V_{in} = \frac{Z_C}{Z_R + Z_C}  = \frac{1/(i\omega C)}{R+1/(i\omega C)}$$
$$ = \frac{1}{i\omega RC + 1} = \frac{1}{1+sRC}$$
$$ H_{CR}(s) = \frac{sRC}{1+sRC} = \frac{1}{1/(sRC)+ 1}$$
