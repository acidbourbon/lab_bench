```python
import numpy as np
from matplotlib import pyplot as plt
from my_utils import *
from sigproc_kit import *
```

```python

Q = 100e-12

delay=10e-9
tau1 = 5e-9
tau2 = 10e-9

t = np.linspace(0,1e-6,10000) # be sure to make not to few steps, if you have a sharp IR
y = detector_signal_function(t, Q, tau1, tau2, delay)

Theta = t>10e-9

ycr = CR_filter(t,0.5*Theta,1,tau2)

yrc = RC_filter(t,ycr,1,tau1)


plt.plot(t*1e9,y*1e3,label="det pulse")
plt.plot(t*1e9,Theta*1e3,label="heaviside")
plt.plot(t*1e9,ycr*1e3,"--" ,label="det pulse after CR")
plt.plot(t*1e9,yrc*1e3,"--", label="det pulse after RC")
plt.xlim(-10,200)
plt.xlabel("ns")
plt.ylabel("mV")
plt.legend()
plt.show()

print("rise_10-90 {:3.3f} ns".format(rise_time(t,y)*1e9))
```
