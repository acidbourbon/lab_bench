```python
import numpy as np
from matplotlib import pyplot as plt
from my_utils import *
from sigproc_kit import *
```

```python
t = np.linspace(0,1e-6,10000) # be sure to make not to few steps, if you have a sharp IR
y = detector_signal_function(t, 100e-12, 5e-9, 10e-9, 10e-9)
```

```python

R = 1e3
C = 10e-12

yrc = RC_filter(t,y,R,C)
ycr = CR_filter(t,y,R,C)


plt.plot(t*1e9,y*1e3,label="det pulse")
plt.plot(t*1e9,yrc*1e3,label="det pulse after RC")
plt.plot(t*1e9,ycr*1e3,label="det pulse after CR")
plt.xlim(-10,200)
plt.xlabel("ns")
plt.ylabel("mV")
plt.legend()
plt.show()
```

```python
##################################################
##                 higher order                 ##
##################################################

R = 1e3
C = 10e-12

yrc = RC_filter(t,y,R,C)
ycr = CR_filter(t,y,R,C)

yrc_2 = RC_filter(t,y,R,C,n=2)
yrc_3 = RC_filter(t,y,R,C,n=3)

ycr_2 = CR_filter(t,y,R,C,n=2)
ycr_3 = CR_filter(t,y,R,C,n=3)

plt.plot(t*1e9,y*1e3,label="det pulse")
plt.plot(t*1e9,yrc*1e3,label="det pulse after RC")
plt.plot(t*1e9,yrc_2*1e3,"--",label="2-pole RC")
plt.plot(t*1e9,yrc_3*1e3,"--",label="3-pole RC")
plt.xlabel("ns")
plt.ylabel("mV")
plt.xlim(-10,200)
plt.legend()
plt.show()

plt.plot(t*1e9,y*1e3,label="det pulse")
plt.plot(t*1e9,ycr*1e3,label="det pulse after CR")
plt.plot(t*1e9,ycr_2*1e3,"--",label="2-pole CR")
plt.plot(t*1e9,ycr_3*1e3,"--",label="3-pole CR")
plt.xlim(-10,200)
plt.xlabel("ns")
plt.ylabel("mV")
plt.legend()
plt.show()
```

```python

plt.plot(t*1e9,integrate_dt(t,  y)*1e9,label="det pulse")
plt.plot(t*1e9,integrate_dt(t,yrc)*1e9,label="det pulse after RC")
plt.plot(t*1e9,integrate_dt(t,ycr)*1e9,label="det pulse after CR")
plt.xlim(-10,200)
plt.xlabel("ns")
plt.ylabel("nV*s")
plt.title("integral")
plt.legend()
plt.show()
```
