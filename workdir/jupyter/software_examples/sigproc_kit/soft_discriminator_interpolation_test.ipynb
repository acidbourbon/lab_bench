```python
import numpy as np
from matplotlib import pyplot as plt
from my_utils import *
from sigproc_kit import *
```

```python
##################################################
##             basic discriminator              ##
##################################################
big_figure()
t = np.linspace(0,100e-9,11) # be sure to make not to few steps, if you have a sharp IR
det = detector_signal_function(t, 100e-12, 10e-9, 20e-9, 10e-9)

noise = np.random.normal(size=len(t))*0.00

y = det + noise

thresh = 50e-3

discr, t1, tot = discriminate(t,y,thresh,interpolate=False)

plt.plot(t*1e9,y*1e3,label="det pulse")
plt.plot(t*1e9,y*0+thresh*1e3,label="threshold")
plt.plot(t*1e9,discr*100-120,label="discr. out")
plt.xlabel("ns")
plt.ylabel("mV")
plt.legend()
plt.show()

print("t1 : {:3.3f} ns, tot : {:3.3f} ns".format(t1*1e9,tot*1e9))
```
