```python
import numpy as np
from matplotlib import pyplot as plt
from sigproc_kit import *
```

```python
x = np.linspace(0,10,1000)
y = (np.sin(10*x)**2)*(np.sin(x/2)**2)
plt.plot(x,y)
```

```python

plt.plot(x,y)
plt.plot(x,fft_gauss_LPF(x,y,sigma=0.2))
```

```python

shy = y-fft_gauss_LPF(x,y,sigma=0.2)

thr = 0.05

plt.plot(x,shy)
plt.plot(x,x*0 + thr)
```

```python
discr, t1_list, tot_list = discriminate(x,shy,thr, multi_hit=True)


plt.plot(x,discr)
```

```python
t2_list = t1_list + tot_list

peak_list = t1_list + tot_list/2
```
