```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
```

```python
# Read Images 
img = mpimg.imread('images/doggo.png') 
  
# Output Images 
plt.imshow(img) 
plt.show()
```

```python
x= (img[:,:,0] + img[:,:,1] + img[:,:,2])/3
plt.imshow(x)
plt.show()
```

```python
m,n = x.shape
X, Y = np.meshgrid(range(m),range(n))
a = np.sin(Y*2*np.pi/10)
plt.imshow(a)
```

```python
x = x+a
plt.imshow(x)
```

```python
from numpy.fft import fft2, ifft2
""" 2D convolution, using FFT"""
fr = fft2(x)
m,n = fr.shape
fr = np.roll(fr, int(-m/2),axis=0)
fr = np.roll(fr, int(-n/2),axis=1)

plt.imshow(np.imag(fr))
plt.show()
plt.imshow(np.log(np.abs(fr)))

plt.show()


#fr2 = fft2(np.flipud(np.fliplr(y)))
#m,n = fr.shape

fr = np.roll(fr, -int(-m/2),axis=0)
fr = np.roll(fr, -int(-n/2),axis=1)

cc = np.real(ifft2(fr))
plt.imshow(np.real(cc))
plt.show()
#cc = np.roll(cc, int(-m/2+1),axis=0)
#cc = np.roll(cc, int(-n/2+1),axis=1)
#return cc
```
