```python
import numpy as np
```

```python
def gen_codebook(bits):
    
    cb = ["0","1"]
    
    for i in range(bits-1):
        cbr = cb[::-1]
        cb = [ "0" + x for x in cb]
        cbr = [ "1" + x for x in cbr]
        cb = cb + cbr
        
    return cb
    
```

```python
gen_codebook(4)
```
