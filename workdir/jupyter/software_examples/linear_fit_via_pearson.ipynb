```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
size = 1000

x = np.random.normal(size=size) + 3
y = 3.14159*x + 0.2*np.random.normal(size=size) + 2
```

```python
from scipy.optimize import curve_fit

def func(x, a, b):
    return a*x + b

p0 = [2,-0.5]
popt, pcov = curve_fit(func, x, y, p0 = p0)

plt.scatter(x,y,alpha=0.5,s=1)
#plt.plot(x,func(x,*p0),"r")
plt.plot(x,func(x,*popt),"r",label="fit")
plt.grid(True)
plt.title("regular linear fit")
plt.show()

print("slope, y0:")
print(popt)
```
