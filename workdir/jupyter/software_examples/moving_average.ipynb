```python
a = np.zeros(100)
```

```python
def movavg(x,width):
    
  if (width%2) == 0:
    print("neee, mach ungerade breite mann !")
    return "bullshit"

  y = np.zeros(len(x))  
  for i in range(0,len(x)):
    for j in range(i-int((width-1)/2),i+int((width-1)/2)):
      if(j < 0):
        j = 0
      if(j > (len(x)-1) ):
        j = len(x)-1
      y[i] += x[j]
    y[i]/=int(width)
  return y
```

```python
def discrete_LPF(x,n):
  y = np.zeros(len(x))
  last_y = 0
  for i in range(0,len(x)):
    new_y = last_y*(n-1)/n + x[i]*1/n
    y[i] = new_y
    last_y = new_y
    
  return y
    
```

```python
plt.scatter(range(0,len(a)),a,label="original",s=0.5)
plt.plot(range(0,len(a)),movavg(a,21),label="filtered",color="orange")
plt.legend()
plt.show()
```

```python
plt.plot(range(0,len(a)),a,label="original")
plt.plot(range(0,len(a)),discrete_LPF(a,10),label="filtered",color="orange")
plt.legend()
plt.show()
```
