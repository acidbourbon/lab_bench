```python
import numpy as np
from matplotlib import pyplot as plt
from my_utils import *
from sigproc_kit import *
from scipy.optimize import curve_fit
```

```python
def func(x,a,b):
    return x*a+b
```

```python
v = np.array([0,.1,.2,.3,.4])
T = np.array([930,949,968,987,1005])*1e-9/100
```

```python
plt.scatter(v,T*1e9)
plt.ylabel("ns")

popt0 = [.1/0.4,0.9]
popt, pcov = curve_fit(func, v, T, p0=popt0)
print(popt)

v_ex = np.linspace(-1.5,0.4,100)

plt.title("the linear range of the varactor circuit")
plt.xlabel("v_bias (V)")
plt.ylabel("T (ns)")
plt.plot(v_ex,func(v_ex,*popt)*1e9)
```
