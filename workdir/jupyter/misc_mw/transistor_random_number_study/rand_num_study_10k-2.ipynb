```python
cov = np.array([0.]*7)
for num in data:
  a = num&1
  for i in range(1,8):
    b = (num & (1<<i))>>i
    cov[i-1] += (float(a)-0.5)*(float(b)-0.5)    
    
cov/=len(data)
```

```python
x= np.array(range(1,8))
plt.plot(x,cov,color="orange")
plt.scatter(x,cov)
#plt.plot(x,0.18*np.exp(-0.45*x))
#plt.ylim(0,0.15)
plt.title("n-nearest neighbor covariance")
```
