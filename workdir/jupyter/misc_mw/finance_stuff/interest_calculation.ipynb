```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# bondora scenario

irate = 1.06

inflation = 1-0.02

start_val = 1000

years = 15

t = np.arange(0,years,0.1)

stale_val = start_val * (inflation **t)


mult = irate **t * inflation **t

t_vals = start_val * mult

plt.plot(t,t_vals)
plt.plot(t,stale_val,label="stale (inflation)")
plt.ylim(0,t_vals[-1]*1.1)

plt.show()
```

```python
# pessimistic mintos scenario

irate = 1.10

start_val = 1000

extraction_loss = 0.25



mult = (irate **t) * (inflation **t)

t_vals_mintos = start_val * mult 
t_vals_mintos_extract = start_val * mult * (1-extraction_loss)


plt.plot(t,t_vals,label="bondora")
plt.plot(t,t_vals_mintos,label="mintos")
plt.plot(t,t_vals_mintos_extract,label="mintos w extract loss")
plt.plot(t,stale_val,label="stale (inflation)")
plt.ylim(0,np.max([t_vals_mintos[-1],t_vals[-1]]))
plt.legend()
plt.show()
```
