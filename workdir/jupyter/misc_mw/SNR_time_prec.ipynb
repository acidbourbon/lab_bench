```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data_v = np.array([250e-6, 500e-6, 1e-3,2e-3])
data_rms = np.array([110e-12, 50e-12, 22e-12,15e-12])


sigma_v = 33e-6
t_rise = 0.8e-9

ana_v = np.linspace(100e-6,10e-3,1000)
ana_rms = t_rise/(ana_v/sigma_v) # t_rise/SNR
```

```python
plt.scatter(data_v/1e-6,data_rms/1e-12,label="data")
plt.plot(ana_v/1e-6,ana_rms/1e-12,"r",label="theory")
#plt.xlim(ana_v[0]/1e-6,ana_v[-1]/1e-6)

plt.legend()
plt.xlabel("signal step (uV)")
plt.ylabel("RMS prec")
plt.xlim(0,2500)
plt.title("assuming intrinsic noise of 33 uV (DC->1GHz)")
plt.show()
```
