```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def rotate(l, x):
  return l[-x:] + l[:-x]

def rotate_array(l,x):
  return np.concatenate((l[-x:],l[:-x]),axis=0)
```

```python
x = np.linspace(0,2*np.pi,2**9+1)
x = x[:-1]
phase = 0.2
u = np.sin(x)
v = np.cos(3*x+phase)
plt.plot(u,v)
```

```python
w = x*0
order = 16
for i in range(1,order,2):
    w += 1/i*np.sin(i*x)

plt.plot(x,w)
```

```python
plt.plot(w,rotate_array(w,5))
```
