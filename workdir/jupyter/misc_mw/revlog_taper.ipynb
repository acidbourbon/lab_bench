```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x= np.linspace(0,1,100)

a = 4.7
b = a
y= np.exp(a*(x-1))
yrev = np.log(x)/b+1

plt.rcParams["figure.figsize"] = (6,6)

plt.plot(x,y,label="audio/log taper")
plt.plot(x,yrev,label="RevLog taper")
plt.plot(x,x,label="lin taper")
plt.scatter([0.5,0.5,0.1,0.9],[0.1,0.9,0.5,0.5],color="red")
plt.xlabel("fraction of full rotation")
plt.ylabel("fraction of full resistance")
plt.legend()
plt.show()
```

```python
x= np.linspace(0,1,100)

a = 6.6
# first solution
b = (np.sqrt(a)*np.sqrt(a+4)-a)/(2*a)
y = 1-(1/(a*(x+b))-b) 

# second solution
b2 = (-a-np.sqrt(a+4)*np.sqrt(a))/(2*a)
y2 = 1-(1/(a*(x+b2))-b2) 


#yrev = np.log(x)/b+1



plt.rcParams["figure.figsize"] = (6,6)

plt.plot(x,y,label="revlog taper")
plt.plot(x,y2,label="log taper")
#plt.plot(x,yrev,label="RevLog taper")
plt.plot(x,x,label="lin taper")
plt.scatter([0.5,0.5,0.1,0.9],[0.1,0.9,0.5,0.5],color="red")
plt.xlabel("fraction of full rotation")
plt.ylabel("fraction of full resistance")
plt.ylim(0,1)
plt.xlim(0,1)
plt.legend()
plt.show()
```
