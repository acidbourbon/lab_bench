```python
C = 100e-12
R1 = 1e3 # serial in 
R2 = 1e3 # par to gnd
R3 = 220e3 # feedback

R12 = 1/(1/R1+1/R2)
f0 = 1/(2*np.pi*C*np.sqrt(R12*R3))
B = 2/(2*np.pi*C*R3)
H0 = R3/(2*R1)

print("f0: {:3.3f}kHz, B: {:3.3f}kHz, H0: {:3.3f}".format(f0/1e3,B/1e3,H0))
print("B/f0 {:3.3f}".format(B/f0))

# at higher frequencies GBWP of amplifier limits the resonance frequency
```
