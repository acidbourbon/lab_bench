```python
test_string="Manchester code always has a transition at the middle of each bit period and may (depending on the information to be transmitted) have a transition at the start of the period also. The direction of the mid-bit transition indicates the data. Transitions at the period boundaries do not carry information. They exist only to place the signal in the correct state to allow the mid-bit transition. "


print(test_string)
```

```python
def all_pair_balanced_codes(n):
  selection = []
  
  #n = word length
  
  n = int(n/2)
  
  for i in range(0,2**(n)):
    #print(i)
    formatstring = "{:0"+str(n)+"b}"
    bitstring = formatstring.format(i)
    bitstring = bitstring.replace("0", "A")
    bitstring = bitstring.replace("1", "B")
    bitstring = bitstring.replace("A", "01")
    bitstring = bitstring.replace("B", "10")
    #onecount = len(bitstring.replace("0",""))
    #print(bitstring)
    #print(onecount)
    #if onecount == n/2:
    selection += [bitstring]
  
  return selection
```

```python
codebook = all_pair_balanced_codes(20)[0:256]
print(codebook)
print(len(codebook))
```

```python
reverse_codebook = {}
for i in range(len(codebook)):
    code = codebook[i]
    reverse_codebook[code] = i
reverse_codebook
```

```python
comma = "01101010101010101010"
```

```python
def string_to_chars(string):
    
    byte_list = []
    
    for char in string:
        byte_list += [ord(char)]
        
    return byte_list
```

```python
def chars_to_string(chars):

    reconstructed_string = ""
    
    for byte in chars:
        reconstructed_string += chr(byte)
        
    return reconstructed_string
```

```python
def enc20(list):
    out_str = ""
    for char in list:
        out_str += codebook[char]
    return out_str
        
```

```python
def dec20(stream):
    char_list = []
    for i in range(int(len(stream)/20)):
        chunk = stream[i*20:(i+1)*20]
        
        if chunk == comma:
            char_list += string_to_chars("<,>")
        else:
            try:
                char_list += [reverse_codebook[chunk]]
            except:
                char_list += [0]
    return char_list
        
```

```python
def barrel_shift(string,n):
    if( n == 0):
        return string
    else:
        m = -n
        return string[m:] + string[0:m]
```

```python
def align(stream):
    pos = stream.find(comma)
    if pos != -1:
        return stream[pos+20:]
    else:
        return ""
        
        
```

```python
##################################################
##         regular insertions of commas         ##
##################################################


def add_comma_every_n_symbols(stream,n):
    out_stream=""
    for i in range(int(int(len(stream)/20)/n)):
        out_stream += comma + stream[i*n*20:(i+1)*n*20] 
        
    return out_stream
```
