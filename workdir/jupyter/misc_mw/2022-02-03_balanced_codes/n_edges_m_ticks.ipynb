```python
def all_words_of_length(n):
  selection = []
  
  for i in range(0,2**n):
    formatstring = "{:0"+str(n)+"b}"
    bitstring = formatstring.format(i)
    onecount = len(bitstring.replace("0",""))
    selection += [bitstring]
  
  return selection
```

```python
def count_rising(word):
    my_word = "0"+word
    
    rising = 0
    for pos in range(0,len(word)):
        if my_word[pos:pos+2] == "01":
            rising += 1
    return rising

def count_ones(word):
    ones = 0
    for pos in range(0,len(word)):
        if word[pos] == "1":
            ones += 1
    return ones
    
```

```python
def find_words(ticks,rising_edges,**kwargs):
    balanced = kwargs.get("balanced",False)
    selection = []
      
    for word in all_words_of_length(ticks):
        
        if balanced:
            if count_ones(word) != ticks/2:
                continue
        if (word[0] == "1") and (word[ticks-1] == "0"):
            if count_rising(word) == rising_edges:
                selection += [word]

    print("ticks = {}, rising edges = {}".format(ticks,rising_edges))
    print("found {} possible words".format(len(selection)))
    return selection
    
```

```python
dummy = find_words(10,1)
dummy = find_words(10,2)
dummy = find_words(10,3)
dummy = find_words(10,4)
dummy = find_words(10,5)
```
