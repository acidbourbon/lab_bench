```python
def all_balanced_codes(n):
  selection = []
  
  for i in range(0,2**n):
    #print(i)
    formatstring = "{:0"+str(n)+"b}"
    bitstring = formatstring.format(i)
    onecount = len(bitstring.replace("0",""))
    #print(bitstring)
    #print(onecount)
    if onecount == n/2:
      selection += [bitstring]
  
  return selection
```

```python
all_balanced_codes(8)
```

```python
def all_pair_balanced_codes(n):
  selection = []
  
  #n = word length
  
  n = int(n/2)
  
  for i in range(0,2**(n)):
    #print(i)
    formatstring = "{:0"+str(n)+"b}"
    bitstring = formatstring.format(i)
    bitstring = bitstring.replace("0", "A")
    bitstring = bitstring.replace("1", "B")
    bitstring = bitstring.replace("A", "01")
    bitstring = bitstring.replace("B", "10")
    #onecount = len(bitstring.replace("0",""))
    #print(bitstring)
    #print(onecount)
    #if onecount == n/2:
    selection += [bitstring]
  
  return selection
```

```python
all_pair_balanced_codes(8)
```

```python
selection = set(all_pair_balanced_codes(8))
```

```python
def check(selection, k):
  #winners = []
  
  verbose = 0
  
  
    
  #for k in candidates:
  if verbose:
    print("#######################")
    print("komma character: {:s}".format(k))
  
  is_winner = True
    
  for i in selection:
    for j in selection:
      # combo character
      combo = i+j
      if verbose:
        print("combo character: {:s}".format(combo))
      if( k in combo ):
        is_winner = False
        if verbose:
          print("match!")
  #if is_winner:
  #  winners += [k]
  return is_winner
```

```python
def check2(selection, k):
  #winners = []
  
  verbose = 0
  
  
    
  #for k in candidates:
  if verbose:
    print("#######################")
    print("komma character: {:s}".format(k))
  
  is_winner = True
    
  for i in selection:
    for j in selection:
      # combo character
      combo = i+j
      if verbose:
        print("combo character: {:s}".format(combo))
      if( k in combo[1:-1] ):
        is_winner = False
        if verbose:
          print("match!")
  #if is_winner:
  #  winners += [k]
  return is_winner
```

```python
new_specials = []
```
