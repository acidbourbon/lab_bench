```python
import numpy as np
from matplotlib import pyplot as plt

from my_utils import *
from IPython.display import clear_output

import copy

from my_solver import *
```

```python
my_grid = grid(-1.5,1.5,.05,-1,1,.05)

q_density = field(my_grid)
e_field = field(my_grid,nd=2)
```

```python
# number of particles
n = 150

c1 = conductor(-1.4,1.4,-0.5,0)
c2 = conductor(-.3,.3,0.2,0.3)

particles = []

for i in range(int(n/2)):
    particles += [qcarrier(-1,c1)]
    particles += [qcarrier(+1,c2)]
```

```python

plot_particles(particles)
plt.xlim(-1.5,1.5)
plt.ylim(-1,1)
plt.show()
```

```python
e_field_from_particles(e_field,particles)

plt.imshow(e_field.matrix[:,:,0])
plt.show()
plt.imshow(e_field.matrix[:,:,1])
plt.show()
plt.quiver(e_field.matrix[:,:,0],e_field.matrix[:,:,1])
plt.show()
```

```python
timesteps = 200

clear_animation_cache()

for t in range(timesteps):
    clear_output(wait=True)
    plot_particles(particles)
    plt.xlim(-1.5,1.5)
    plt.ylim(-1,1)
    plt.title("electron positions")
    save_animation_frame(plt)
    plt.show()
    particles = move_particles(particles,strength=2e-5,friction=0.5,granularity=1e-4,vmax=1e-2)
```
