```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data_v   = np.array([350e-6 ,500e-6, 1e-3    , 2e-3    ,4e-3     ,10e-3 ,20e-3 ,50e-3    ,100e-3  ,400e-3   ])
data_rms = np.array([101e-12,53e-12, 28.5e-12, 22.3e-12, 18.1e-12,20e-12,20e-12, 18.7e-12,22.5e-12, 20.7e-12])


```

```python
plt.scatter(data_v/1e-3,data_rms/1e-12,label="data")
#plt.plot(ana_v/1e-6,ana_rms/1e-12,"r",label="theory")
#plt.xlim(ana_v[0]/1e-6,ana_v[-1]/1e-6)

plt.legend()
plt.grid(True)
plt.xlabel("signal step (mV)")
plt.ylabel("RMS prec")
#plt.xlim(0,2500)
plt.ylim(0,110)
plt.xscale("log")
plt.title("RMS time precision CPS+PADIWA3")
plt.show()
```

```python
sigma_v = 50e-6
t_rise = 0.5e-9
const_jitter = 20e-12

ana_v = np.exp(np.linspace(np.log(300e-6),np.log(400e-3),1000))
ana_rms = t_rise/(ana_v/sigma_v) # t_rise/SNR
ana_rms = np.sqrt( ana_rms*ana_rms
    + const_jitter*const_jitter)

plt.scatter(data_v/1e-3,data_rms/1e-12,label="data")
plt.plot(ana_v/1e-3,ana_rms/1e-12,"r",label="theory")
#plt.xlim(ana_v[0]/1e-6,ana_v[-1]/1e-6)

plt.legend()
plt.grid(True)
plt.xlabel("signal step (mV)")
plt.ylabel("RMS prec")
#plt.xlim(0,2500)
plt.ylim(0,110)
plt.xscale("log")
plt.title("RMS time precision CPS+PADIWA3")
plt.show()
```
