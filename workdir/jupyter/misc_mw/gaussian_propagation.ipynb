```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def my_std(data):
  mu = np.mean(data)
  thesum = 0
  for i in range(0,len(data)):
    deviation = data[i]-mu
    thesum += deviation**2
  var = thesum/len(data)

  return np.sqrt(var)

def odev(data):
  mu = np.mean(data)
  thesum = 0
  for i in range(0,len(data)):
    deviation = data[i]-mu
    thesum += np.abs(deviation)
  absvar = thesum/len(data)

  return absvar
    
    
```
