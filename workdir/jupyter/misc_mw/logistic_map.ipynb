```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib notebook
```

```python
def end_states(x0,r,N,n):
  x = x0
  Y = []
  for i in range(0,N):
    x = r*x*(1-x)
    if abs(x) < 10:
      if i > (N-n):
        Y += [x]
    if abs(x) > 1e6:
      break
  return Y
```

```python
x0 = 0.5
N = 512
n = 64

#end_states(x0,r,N,n)


for r in np.linspace(2.7,4,3000):
  s = end_states(x0,r,N,n)
  plt.scatter([r]*len(s),s,color="red",s=0.01,alpha=0.5)

plt.xlabel("R")
plt.ylabel("X_n")
plt.show()
```
