```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib notebook

import random
import pandas as pd
import math
```

```python
np.random.random()
```

```python
x = []
N = 10000
coins = 20
for i in range(0,N):
  c = []
  for j in range(0,coins):
    c += [ (np.random.random() < 0.2) ]
  x += [ c ]

#x
```

```python
# outcomes coded by numbers
# [0,0] -> 0
# [0,1] -> 1
# [1,0] -> 2
# [1,1] -> 3

z = []
for i in range(0,len(x)):
    z += [  np.array(x[i]).sum()  ]
#z
```

```python

xx = pd.Series(z)

#xx.plot.hist(grid=True, bins=10, color='#607c8e', xlim=(-1,5))
xx.plot.hist(grid=True, bins=coins*2, color='#607c8e', range=(0,coins))#, xlim=(75,125))
plt.title('blah')
plt.xlabel('sum of heads in {:d} coinflips'.format(N))
plt.ylabel('freq')
plt.grid(axis='y', alpha=0.75)
```

```python
def binom(n,k,p):
    q = 1-p
    return math.factorial(n)/(math.factorial(n-k)*math.factorial(k)) * p**k * q**(n-k)
    
```
