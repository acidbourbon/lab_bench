```python
from sigproc_kit import *
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.linspace(-5,5,1000)

y=2*(gauss(x) - gauss(x,sigma=0.5)) + 3

plt.plot(x/20,y,label="signal")
plt.plot(x/20,y*0+3,label="baseline")
plt.plot(x/20,y*0+3.1,"--",label="threshold")
plt.ylabel("voltage (V)")
plt.xlabel("time (s)")
plt.legend()
plt.show()
```
