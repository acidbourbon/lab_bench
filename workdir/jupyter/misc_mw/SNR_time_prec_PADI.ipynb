```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data_v   = np.array([ 150e-6, 250e-6/1.41,  250e-6, 250e-6*1.41, 500e-6, 500e-6*1.41,   1e-3, 1.41e-3,   2e-3,
                    3e-3])
data_rms = np.array([150e-12,     110e-12,   87e-12,      78e-12, 55e-12,      56e-12, 49e-12,  46e-12, 40e-12,
                    44e-12])


sigma_v = 27e-6
t_rise = 0.8e-9

ana_v = np.linspace(100e-6,10e-3,1000)
ana_rms = t_rise/(ana_v/sigma_v) # t_rise/SNR
```

```python
intrinsic_error = 40e-12


plt.scatter(data_v/1e-6,data_rms/1e-12,label="data")
plt.plot(ana_v/1e-6,ana_rms/1e-12,"r",label="theory SNR")

gaussian = np.sqrt(ana_rms*ana_rms + intrinsic_error*intrinsic_error)
plt.plot(ana_v/1e-6,gaussian/1e-12,"orange",label="theory² + 40ps²")

plt.legend()
plt.xlabel("signal step (uV)")
plt.ylabel("RMS prec (ps)")
plt.xlim(0,3200)
plt.title("time precision vs input amplitude\n assuming t_rise = 0.8ns\n RMS input noise = 27uV")
plt.show()
```
